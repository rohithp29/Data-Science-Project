```python
df=df.drop(columns=['CustomerID','Genre'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i, init='k-means++',random_state=42)
  kmeans.fit(x_scaled)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='g')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=42)
df['Cluster'] = kmeans.fit_predict(x_scaled)
print(df['Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_pca[:, 0], y=x_pca[:, 1], hue=df['Cluster'], palette='Set2', s=100, alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Customer Segmentation Cluster")
plt.legend(title="Cluster")
plt.show()
```

```python
sns.pairplot(df, hue='Cluster', palette='viridis')
plt.suptitle("Pairplot of Customer Clusters", y=1.02)
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(x_scaled)

centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_pca[:, 0], y=x_pca[:, 1], hue=df['Cluster'], palette='Set2', s=100, alpha=0.7)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], s=300, c='black', marker='X', label='Centroids')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Customer Segmentation with Centroids")
plt.legend(title="Cluster")
plt.show()
```
